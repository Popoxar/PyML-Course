```python
data = [[['koudak', 'nojavan']], [['koudak', 'siasi']], [['siasi']]]
  
df = pd.DataFrame(data, columns = ['Genre']) 

df 
```

```python
cnt={}
def count_terms(term):
    for t in term:
        if t in list(cnt.keys()):
            cnt[t] += 1
        else:
            cnt[t] = 1

df.Genre.apply(count_terms)
```

```python
films=pd.read_csv('FilmsFinal.csv')
```

```python
films.head()
```

```python
from sklearn.cluster import KMeans
```

```python
sales=films['جمع فروش (تومان)']
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

X = np.reshape(sales.values, (-1, 1))

```

```python
X
```

```python
ms = MeanShift(bandwidth=None, bin_seeding=True)
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
print(labels)
```

```python
from hdbscan import HDBSCAN

clusterer = HDBSCAN(min_cluster_size=3)
cluster_labels = clusterer.fit_predict(X)

best_cluster = clusterer.exemplars_[cluster_labels[X.argmax()]].ravel()
print(best_cluster)
```
