### str.format()

```
# All following codes have same output

statement = '{0} is an interpreted, {1}, general-purpose programming {2}.'.format('Python','high-level','language')
print(statement)

statement = '{} is an interpreted, {}, general-purpose programming {}.'.format('Python','high-level','language')
print(statement)

statement = '{lang} is an interpreted, {level}, general-purpose programming {thing}.'.format(level='high-level', thing='language', lang='Python')
print(statement)

lang='Python'
level='high-level'
thing='language'
statement = f'{lang} is an interpreted, {level}, general-purpose programming {thing}.'
print(statement)
```

```
# decimal (.) precision of 3 for float '0.333'
print('{0:.3f}'.format(1.0/3))

# fill with underscores (_) with the text centered
# (^) to 11 width '___hello___'
print('{0:_^11}'.format('hello'))

# keyword-based 'Guido van Rossum creates Python'
print('{name} creates {thing}'.format(name='Guido van Rossum', thing='Python'))
```

### int

```
# int can be -1, 0, 9223372036854775807
my_int = 6
print('value: {}, type: {}'.format(my_int, type(my_int)))

# int max value in Python 3
import sys
max = sys.maxsize
min = -sys.maxsize -1
print(f"min value in Python 3 is {min} and max value is: {max}")
```

### float

```
my_int=100
my_float = float(my_int)
print('value: {}, type: {}'.format(my_float, type(my_float)))
```

```
# division of ints produces float
print(1 / 1)
print(6 / 5)
```

```
# Be aware of the binary floating-point pitfalls
val = 0.1 + 0.1 + 0.1
print(val == 0.3)
print(val)
```

### decimal.Decimal


```
from decimal import Decimal
```

```
from_float = Decimal(0.1)
from_str = Decimal('0.1')
print('from float: {}\nfrom string: {}'.format(from_float, from_str))
```

```
my_decimal = Decimal('0.1')
sum_of_decimals = my_decimal + my_decimal + my_decimal
print(sum_of_decimals == Decimal('0.3'))
```

```
# Operator precedence in calculations

print(1 + 2**2 * 3 / 6) # 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```
